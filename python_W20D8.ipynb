```python
import pandas as pd
import matplotlib.pyplot as plt

# Carico il file Excel
file_path = '/Users/ildon91/Desktop/CORSO EPICODE/M4/dati-finalassignment/dati-canzoni-spotify-sanremo-1951-2023.xlsx'
xls = pd.ExcelFile(file_path)

# Carico i dati dal foglio "Sheet1"
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Visualizzo le prime righe del dataframe
print(df.head())

# Rimuovo le righe duplicate
df_cleaned = df.drop_duplicates()

# Verifico che i duplicati siano stati rimossi
print("Righe duplicate rimosse:", df_cleaned.duplicated().sum())

# 1. Evoluzione della Popolarit√† delle Canzoni di Sanremo nel Tempo

# Cosa dimostra:
# Come la popolarit√† delle canzoni di Sanremo √® cambiata dal 1951 al 2023.
# Le tendenze generali di crescita o declino nella popolarit√† delle canzoni.
# Eventuali picchi o cali significativi in specifici anni, che possono essere correlati a eventi storici o cambiamenti nel festival.

# Raggruppo i dati per anno e calcolo la media della popolarit√†
popularity_over_time = df_cleaned.groupby('Anno_ultimo_rilascio')['Popolarit√†'].mean()

# Visualizzo la popolarit√† nel tempo
plt.figure(figsize=(12, 6))
popularity_over_time.plot(kind='line')
plt.title('Popolarit√† delle Canzoni di Sanremo nel Tempo')
plt.xlabel('Anno')
plt.ylabel('Popolarit√† Media')
plt.grid(True)
plt.show()
```

```python
# 2. Distribuzione della Durata delle Canzoni nel Tempo

# Cosa dimostra:
# Come la durata delle canzoni di Sanremo √® cambiata nel tempo.
# Le tendenze verso canzoni pi√π lunghe o pi√π corte in specifici periodi.
# L'evoluzione delle preferenze di durata da parte degli artisti e dei produttori del festival.

# Raggruppo i dati per anno e calcolare la durata media
duration_over_time = df_cleaned.groupby('Anno_ultimo_rilascio')['Durata (min)'].mean()

# Visualizzo la durata nel tempo
plt.figure(figsize=(12, 6))
duration_over_time.plot(kind='line', color='orange')
plt.title('Durata Media delle Canzoni di Sanremo nel Tempo')
plt.xlabel('Anno')
plt.ylabel('Durata Media (minuti)')
plt.grid(True)
plt.show()
```

```python
# 3. Analisi degli Artisti Pi√π Ricorrenti

# Cosa dimostra:
# Gli artisti che hanno partecipato pi√π frequentemente al Festival di Sanremo.
# L'importanza e la presenza ricorrente di certi artisti nel panorama musicale italiano.
# La correlazione tra la frequenza delle partecipazioni e la popolarit√† degli artisti.


# Conto il numero di canzoni per ogni artista
artist_counts = df_cleaned['Artista'].value_counts().head(10)

# Visualizzo gli artisti pi√π ricorrenti
plt.figure(figsize=(12, 6))
artist_counts.plot(kind='bar', color='green')
plt.title('Artisti Pi√π Ricorrenti a Sanremo')
plt.xlabel('Artista')
plt.ylabel('Numero di Canzoni')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()



```
